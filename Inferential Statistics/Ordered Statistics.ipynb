##Probability Distribution of Sample Range
To find the probability distribution of the sample range
ùëÖ, we need to work with the distribution of the difference between the maximum and minimum order statistics from a sample. Let
ùëã(1) be the minimum and ùëã(ùëõ) be the maximum order statistic in a sample of size ùëõ.
The range ùëÖ is defined as:

ùëÖ = ùëã(ùëõ) ‚àí ùëã (1)


```python
'''
The code simulates 1000 random samples from a uniform distribution and calculates the range of each sample.
It then plots the distribution of these ranges to analyze how often certain ranges occur when drawing random numbers from a uniform distribution between 0 and 1.
This type of analysis can be useful for understanding the variability of data drawn from a uniform distribution
In leyman language it visualizes the probability distribution of the sample range
'''

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style='whitegrid')

# Parameters
n_samples = 1000  # Number of samples to generate
sample_size = 30  # Size of each sample
low = 0           # Lower bound of the uniform distribution
high = 1          # Upper bound of the uniform distribution

# Generate random samples and calculate sample ranges
sample_ranges = []

for _ in range(n_samples):
    sample = np.random.uniform(low, high, sample_size)
    sample_range = np.max(sample) - np.min(sample)
    sample_ranges.append(sample_range)

# Convert to numpy array for easier manipulation
sample_ranges = np.array(sample_ranges)

# Create a histogram of sample ranges
plt.figure(figsize=(10, 6))
sns.histplot(sample_ranges, bins=30, kde=True, color='skyblue', stat='density')

# Customize the plot
plt.title('Probability Distribution of Sample Range')
plt.xlabel('Sample Range')
plt.ylabel('Density')
plt.xlim(0, high - low)  # Set x-axis limits
plt.grid(True)

# Show the plot
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
n_samples = 1000  # Number of samples
sample_size = 30   # Size of each sample
mu = 0             # Mean of the normal distribution
sigma = 1          # Standard deviation of the normal distribution

# Step 1: Simulate Sample Data
sample_ranges = []
for _ in range(n_samples):
    sample = np.random.normal(mu, sigma, sample_size)
    sample_range = np.max(sample) - np.min(sample)  # Calculate the sample range
    sample_ranges.append(sample_range)

# Step 2: Visualize the Distribution
plt.figure(figsize=(12, 6))
sns.histplot(sample_ranges, bins=30, kde=True, stat='density', color='skyblue', edgecolor='black')
plt.title('Probability Distribution of Sample Range')
plt.xlabel('Sample Range')
plt.ylabel('Density')
plt.grid()
plt.show()
```

The probability density function (PDF) of the minimum order statistic \($X_{(1)}$\) and maximum order statistic \($X_{(n)}$\) for a sample of size \(n\) from a continuous distribution with $PDF(f(x))$ and cumulative distribution function $(CDF)(F(x))$ is given by:

\
**Minimum order statistic:**\
\
$F_{X_{(1)}}(x) = 1\ - (1 - F(x))^{n - 1} \$


\
**Maximum order statistic :**\
\
$F_{X_{(n)}}(x) = n \cdot F(x)^{n - 1} \cdot f(x)$


```python
'''
Creating a visualization to understand the order statistics
'''

import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=25)

# Calculate order statistics
order_statistics = np.sort(data)

# Identify min and max
min_value = np.min(order_statistics)
max_value = np.max(order_statistics)
min_index = np.argmin(order_statistics)
max_index = np.argmax(order_statistics)

# Create a figure
plt.figure(figsize=(16, 6))

# Plotting the order statistics
plt.plot(order_statistics, marker='o', linestyle='-', color='b', markersize=5)
plt.title('Order Statistics with Min and Max Values')
plt.xlabel('Order (Index)')
plt.ylabel('Value')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add x-axis
plt.xticks(ticks=np.arange(len(order_statistics)), labels=np.arange(1, len(order_statistics) + 1))

# Annotate min and max points
plt.text(min_index, min_value, f'Min: {min_value:.2f}',
         fontsize=10, ha='right', color='red', fontweight='bold')
plt.text(max_index, max_value, f'Max: {max_value:.2f}',
         fontsize=10, ha='right', color='green', fontweight='bold')

# Show the plot
plt.show()
```
