#Random Sampling
let X1, X2..., X3 be a  **i.i.d.** random variable each having the same prob distribution f(x) then we can say that (X1, X2,...,Xn) is a random sample from popluation with distribution f(x) & the joint distribution of X1, X2,..Xn is defined as

```
f(x1, x2,...,xn)= f(x1)*f(x2)...f(xn)
```

##Random Sampling in Discrete Random Variables

````python
#Bernoulli Distribution
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.5  # Probability of success
n_samples = 1000  # Number of samples

# Generate Bernoulli samples
bernoulli_samples = np.random.binomial(n=1, p=p, size=n_samples)

# Visualization
plt.figure(figsize=(8, 4))
plt.hist(bernoulli_samples, bins=2, edgecolor='black', align='left', rwidth=0.6)
plt.xticks([0, 1])
plt.title('Bernoulli Distribution Samples')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.show()
````

````python
#Binomial Distribution

# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success
n_samples = 1000  # Number of samples

# Generate Binomial samples
binomial_samples = np.random.binomial(n=n, p=p, size=n_samples)

# Visualization
plt.figure(figsize=(8, 4))
plt.hist(binomial_samples, bins=range(n+2), edgecolor='black', align='left', rwidth=0.8)
plt.title(f'Binomial Distribution Samples (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
````

````python
#Geometric Distribution

# Parameters
p = 0.3  # Probability of success
n_samples = 1000  # Number of samples

# Generate Geometric samples
geometric_samples = np.random.geometric(p=p, size=n_samples)

# Visualization
plt.figure(figsize=(8, 4))
plt.hist(geometric_samples, bins=range(1, max(geometric_samples)+1), edgecolor='black', align='left', rwidth=0.8)
plt.title(f'Geometric Distribution Samples (p={p})')
plt.xlabel('Number of Trials until First Success')
plt.ylabel('Frequency')
plt.xlim(0, 20)  # Limit x-axis for better visualization
plt.show()
````

````python
#Poisson Distribution
# Parameters
lambda_poisson = 4  # Average rate
n_samples = 1000  # Number of samples

# Generate Poisson samples
poisson_samples = np.random.poisson(lam=lambda_poisson, size=n_samples)

# Visualization
plt.figure(figsize=(8, 4))
plt.hist(poisson_samples, bins=range(0, max(poisson_samples)+1), edgecolor='black', align='left', rwidth=0.8)
plt.title(f'Poisson Distribution Samples (={lambda_poisson})')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.show()
````

##Continuous Random Variables Sampling

````python
#Uniform Distribution

# Parameters
a = 0  # Lower bound
b = 1  # Upper bound
n_samples = 1000  # Number of samples

# Generate Uniform samples
uniform_samples = np.random.uniform(low=a, high=b, size=n_samples)

# Visualization
plt.figure(figsize=(8, 4))
plt.hist(uniform_samples, bins=20, edgecolor='black', density=True, alpha=0.7)
plt.title('Uniform Distribution Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
````

````python
#Exponential Distribution

# Parameters
lambda_expo = 1.5  # Rate parameter
n_samples = 1000  # Number of samples

# Generate Exponential samples
exponential_samples = np.random.exponential(scale=1/lambda_expo, size=n_samples)

# Visualization
plt.figure(figsize=(8, 4))
plt.hist(exponential_samples, bins=30, edgecolor='black', density=True, alpha=0.7)
plt.title(f'Exponential Distribution Samples (={lambda_expo})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
````

````python
#Normal Distribution

# Parameters
mu = 0  # Mean
sigma = 1  # Standard deviation
n_samples = 1000  # Number of samples

# Generate Normal samples
normal_samples = np.random.normal(loc=mu, scale=sigma, size=n_samples)

# Visualization
plt.figure(figsize=(8, 4))
plt.hist(normal_samples, bins=30, edgecolor='black', density=True, alpha=0.7, label='Sampled Data')

# Overlay the theoretical PDF
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)
plt.plot(x, pdf, 'r', label='Theoretical PDF')

plt.title('Normal Distribution Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
````
